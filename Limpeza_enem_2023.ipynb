```python
df = df.dropna(subset=[
    "NU_NOTA_LC",
    "NU_NOTA_MT",
    "NU_NOTA_CH",
    "NU_NOTA_CN"
])
```

```python
drop = [
    'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT',
    'Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012', 'Q013', 'Q014', 'Q015',
    'Q016', 'Q017', 'Q018', 'Q019', 'Q020', 'Q021', 'Q022', 'Q023', 'Q024',
    'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5',
    'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC',
       'TX_RESPOSTAS_MT', 'TP_LINGUA', 'TX_GABARITO_CN', 'TX_GABARITO_CH',
       'TX_GABARITO_LC', 'TX_GABARITO_MT','CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC',
       'CO_PROVA_MT', 'TP_ENSINO', 'TP_SIT_FUNC_ESC', 'SG_UF_PROVA', 'CO_UF_ESC', 'CO_UF_PROVA', 'TP_ANO_CONCLUIU', 'CO_MUNICIPIO_ESC', 'CO_MUNICIPIO_PROVA', 'TP_ST_CONCLUSAO'
]


df = df.drop(columns=drop, errors='ignore')
```

```python


mask = (
    (df["NU_NOTA_CH"] == 0) |
    (df["NU_NOTA_MT"] == 0) |
    (df["NU_NOTA_LC"] == 0) |
    (df["NU_NOTA_CN"] == 0)
)


df = df[~mask].copy()

df["NOTA_MEDIA"] = (
    df["NU_NOTA_CH"] +
    df["NU_NOTA_MT"] +
    df["NU_NOTA_LC"] +
    df["NU_NOTA_CN"] +
    df["NU_NOTA_REDACAO"]
) / 5
```

```python
mapa_q025 = {
    'A': "Sem internet",
    'B': 'Com internet'
}
df['Q025'] = df['Q025'].replace(mapa_q025)
```

```python
map_tipo_escola = {
    1: "Nao Respondeu",
    2: "Publica",
    3: "Privada"
}
df['TP_ESCOLA'] = df['TP_ESCOLA'].replace(map_tipo_escola)
```

```python
mapa_loc_escola = {
    1: "Urbana",
    2: "Rural"
}
df['TP_LOCALIZACAO_ESC'] = df['TP_LOCALIZACAO_ESC'].replace(mapa_loc_escola)
```
